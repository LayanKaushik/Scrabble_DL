```python
class Rack:
    """
    The player's letters. Each turn a player will draw up to 7 tiles (proportionate to how many they played last turn)
    The play selects which letters to play
    """
    def __init__(self, bag):
        # Creates rack and draws the intial 7 tiles
        self.rack = []
        self.bag = bag
        self.initialize()
    
    def add_to_rack(self):
        # Gives the player a letter from the bag
        self.rack.append(self.bag.pop())
    
    def remove_from_rack(self, tile):
        #Removes a tile from the rack (for example, when a tile is being played).
        self.rack.remove(tile)

    def initialize(self):
        # Gives the player the first 7 letters
        for i in range(7):
            self.add_to_rack()
    
    def rack_length(self):
        return len(self.rack)
    
    def fill_rack(self):
        # Refills the players rack to 7 letters
        while self.rack_length() < 7 and self.bag_tiles() >0:
            self.add_to_rack()

    def draw_random_tile(self):
        if len(self.tiles) > 0:  # ensure there are tiles in the bag
            return self.tiles.pop(random.randint(0, len(self.tiles) - 1))  # remove a random tile from the bag and return it
        else:
            print("The bag is empty.")
            return None
```

```python
class Tile:
    """
    Class that allows for the creation of a tile. Initializes using an uppercase string of one letter,
    and an integer representing that letter's score.
    """
    def __init__(self, letter, letter_values):
        #Initializes the tile class. Takes the letter as a string, and the dictionary of letter values as arguments.
        self.letter = letter.upper()
        if self.letter in letter_values:
            self.score = letter_values[self.letter]
        else:
            self.score = 0

    def get_letter(self):
        #Returns the tile's letter (string).
        return self.letter

    def get_score(self):
        #Returns the tile's score value.
        return self.score

    def add_tile(self, tile):
        self.tiles.append(tile)
```
